```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
chromedriverPath = r'C:\Users\driver\chromedriver.exe'
# Create a Service object with the path to chromedriver
chromeService = webdriver.chrome.service.Service(chromedriverPath)
# Start the service
chromeService.start()
# Create the WebDriver using the Service object
driver = webdriver.Chrome(service=chromeService)
```

```python
url = 'https://www.psychologytoday.com/us/psychiatrists/ny/new-york?'  
driver.get(url)
```

```python
start_page = 1  # The first page you want to scrape
end_page = 3    # The last page you want to scrape

# Create an empty set to store unique links
unique_links = set()

for page in range(start_page, end_page + 1):
    # Find all elements with a specific class
    wait = WebDriverWait(driver, 10)
    elements_with_class = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'results-row.top-divider')))

    # Iterate through the elements and extract links
        # Extract the first link from each element with the class
    for element in elements_with_class:
        try:
        # Extract the first link within the element
            first_link = element.find_element(By.TAG_NAME, 'a').get_attribute('href')

                # Check if the link is not already in the set
            if first_link not in unique_links:
            # Add the first link to the set
                unique_links.add(first_link)
        except Exception as e:
            print(f"An error occurred while extracting links: {str(e)}")

    # Check if there is a "Next" button and if it's not the last page, click it  
    if page < end_page:

        try:
            # Wait for the element to be clickable
            wait = WebDriverWait(driver, 10)
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__layout"]/div/div[3]/div/div[5]/div[23]/div/a[11]')))
            if next_button.is_enabled():
                next_button.click()
        except Exception as e:
            print(f"An error occurred while clicking the 'Next' button: {str(e)}")
    else:
        break
```

```python
# Convert the set of unique links to a list
all_links = list(unique_links)
```

```python
# Create empty lists to store the scraped data
data_list1 = []
h1_elements = []
h2_elements = []
for link in all_links:
    driver.get(link)  # Visit each link in the list

    try:
        # Find and extract the <h1> element with a specific class
        h1_element = driver.find_element(By.CLASS_NAME, 'heading-element.profile-title.heading-element-2')
        h1_text = h1_element.text
    except Exception as e:
        print(f"An error occurred while scraping the <h1> element: {str(e)}")

    try:
        # Find and extract the <h2> element with a specific class
        h2_element = driver.find_element(By.CLASS_NAME, 'profile-suffix-heading')
        h2_text = h2_element.text
    except Exception as e:
        print(f"An error occurred while scraping the <h2> element: {str(e)}")
    try:
        # Find and extract the <h2> element with a specific class
        h3_element = driver.find_element(By.CLASS_NAME, 'row')
        h3_text = h3_element.text
    except Exception as e:
        print(f"An error occurred while scraping the <h2> element: {str(e)}")
            # Append the data (link, H1, and H2) to the list
    data_list1.append((link, h1_text, h2_text, h3_text))
```
